```python
ticker='AAPL'

fetch=fetch_basic_data.StockData(ticker)
info=fetch.get_stockinfo()
df=pd.DataFrame(info)


address=str(df.loc['address1']).split('\n')[0].split('    ')[1], str(df.loc['country']).split('\n')[0].split('    ')[1]
sector=str(df.loc['sector']).split('\n')[0].split('    ')[1]
website=str(df.loc['website']).split('\n')[0].split('    ')[1]
summary=str(df.loc['longBusinessSummary']).split('\n')[0].split('    ')[1]

dict={'Address':address, 'Sector':sector, 'Website':website}
df=pd.DataFrame(dict)
df=df.T

```

### Function for retreiving basic data ( Cleaned )

```python
def str_df(loc_str):
    string=str(df.loc[loc_str])
    string = string.split('\n')[0].split('    ')[1]
    
    return string



str_df('website')
```

### Pricing data retreival 

```python
# Gets yesterdays closing price

def get_yesterday_price():
    pricing=fetch.get_daily_stock()
    y_price=round(float(pricing['Close'][-1]),2)
    return y_price
```

```python
# gets monthly range:
def get_monthly_range():
    pricing=fetch.get_daily_stock()
    pricing=round(pricing['Close'],2)
    min=pricing.tail(30).min()
    max=pricing.tail(30).max()
    print(f'The monthly range is {min} - {max}')
```

```python
# gets yearly range:
def get_yearly_range():
    pricing=fetch.get_daily_stock()
    pricing=round(pricing['Close'],2)
    min=pricing.tail(255).min()
    max=pricing.tail(255).max()
    print(f'The yearly range is {min} - {max}')
```

```python
# Plot for the stock

def insert_plot():
    pricing=fetch.get_daily_stock()
    pricing['Close'].plot()
    plt.xlabel('Date')
    plt.ylabel(f'Price of {ticker}')
    plt.grid()
    plt.show() 
```
