```python
import yfinance as yf

def calculate_optimal_options(symbol, expiration_date):
    # Create a Ticker object
    ticker = yf.Ticker(symbol)
    
    # Fetch the options chain for the given expiration date
    options_chain = ticker.option_chain(expiration_date)
    
    # Get the put options
    puts = options_chain.puts
    
    # Define criteria for optimal options (e.g., strike price close to current price, low premium)
    current_price = ticker.history(period='1d')['Close'].iloc[-1]
    optimal_puts = puts[(puts['strike'] <= current_price * 1.1) & (puts['strike'] >= current_price * 0.9)]
    
    # Sort by premium (ask price) to find the cheapest options
    optimal_puts = optimal_puts.sort_values(by='ask')
    
    return optimal_puts

# Example usage
symbol = 'AAPL'
expiration_date = ticker.options[-1]
optimal_options = calculate_optimal_options(symbol, expiration_date)
optimal_options
```
